```python
# ---------------- Retail Sales Data Analysis ---------------- #
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Create Sample Dataset ---
sales_data = {
    'Date': ['2024-01-02','2024-01-03','2024-01-04','2024-01-05','2024-01-06','2024-01-07'],
    'Product': ['Laptop','Mobile','Tablet','Headphones','Charger','Mobile'],
    'Region': ['South','North','East','West','South','East'],
    'Quantity': [5,10,6,12,20,15],
    'Sales': [2500,4000,1800,600,400,6000]
}

df = pd.DataFrame(sales_data)
df.to_csv('retail_sales.csv', index=False)
print("✅ retail_sales.csv file created\n")

# --- 2. Load Data & Overview ---
data = pd.read_csv('retail_sales.csv')
print(data.head(), "\n")

# --- 3. Basic Cleaning ---
data.dropna(inplace=True)
data['Sales'] = data['Sales'].astype(float)

# --- 4. Insights ---
print("Total Sales:", data['Sales'].sum())
print("\nTop Products:\n", data.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(5))
print("\nSales by Region:\n", data.groupby('Region')['Sales'].sum())

# --- 5. Visualization ---
plt.figure(figsize=(8,5))
data.groupby('Product')['Sales'].sum().sort_values(ascending=False).plot(kind='bar',color='teal')
plt.title('Top Products by Sales')
plt.xlabel('Product'); plt.ylabel('Sales'); plt.tight_layout(); plt.show()

plt.figure(figsize=(6,4))
data.groupby('Region')['Sales'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.title('Sales Distribution by Region'); plt.ylabel(''); plt.tight_layout(); plt.show()
```
